#Text Extraction From PDF

```python
from PyPDF2 import PdfReader

def extract_text_from_pdf(file_path):
    text = ""
    with open(file_path, 'rb') as file:
        pdf_reader = PdfReader(file)
        num_pages = len(pdf_reader.pages)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

file_path = r"C:\Users\tarak\Downloads\Antiragging_form.pdf"
extracted_text = extract_text_from_pdf(file_path)
print(extracted_text)
```

```python
extract_text_from_pdf=NER(extracted_text)
```

```python
for word in extract_text_from_pdf.ents:
    print(word.text,word.start_char, word.end_char,word.label_)
```

```python
displacy.render(extract_text_from_pdf,style="ent",jupyter=True)
```

#Text Extraction From Docx

```python
from docx import Document

def extract_text_from_docx(file_path):
    doc = Document(file_path)
    text = ""
    for paragraph in doc.paragraphs:
        text += paragraph.text + "\n"
    return text

file_path = r"C:\Users\tarak\Downloads\springer reviewed.docx"
extracted_text_from_doc = extract_text_from_docx(file_path)
print(extracted_text_from_doc)
```

```python
extracted_text_from_doc=NER(extracted_text_from_doc)
```

```python
for word in extracted_text_from_doc.ents:
    print(word.text,word.start_char, word.end_char,word.label_)
```

```python
displacy.render(extracted_text_from_doc,style="ent",jupyter=True)
```

```python
import requests
from bs4 import BeautifulSoup

# Specify the URL of the website you want to scrape
url ="https://gmrit.edu.in/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the HTML elements that contain the titles of articles
    # (Inspect the HTML structure of the website to identify the appropriate tags and attributes)
    article_titles = soup.find_all("h2", class_="article-title")

    # Extract and print the titles
    for title in article_titles:
        print(title.text)
else:
    print("Failed to retrieve the web page. Status code:", response.status_code)
```
